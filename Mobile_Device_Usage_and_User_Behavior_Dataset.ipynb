```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset with a raw string to avoid unicode errors
data_path = r'D:\A Study\Python\Data Visualization with Python\user_behavior_dataset.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Let's visualize the distribution of App Usage Time (min/day)
plt.figure(figsize=(10, 6))
sns.histplot(df['App Usage Time (min/day)'], bins=30, kde=True)
plt.title('Distribution of App Usage Time (min/day)')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Boxplot of App Usage Time by User Behavior Class
plt.figure(figsize=(12, 6))
sns.boxplot(x='User Behavior Class', y='App Usage Time (min/day)', data=df)
plt.title('App Usage Time by User Behavior Class')
plt.xlabel('User Behavior Class')
plt.ylabel('App Usage Time (min/day)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Mobile Device Usage Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Screen On Time (hours/day)', y='App Usage Time (min/day)', data=df, hue='Gender')
plt.title('Screen On Time vs. App Usage Time')
plt.xlabel('Screen On Time (hours/day)')
plt.ylabel('App Usage Time (min/day)')
plt.legend(title='Gender')
plt.grid()
plt.show()
```
